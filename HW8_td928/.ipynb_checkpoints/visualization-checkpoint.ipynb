```python
import os
import sys
import pandas as pd
import pylab as pl
import scipy as sp
from scipy import stats
import numpy as np
import statsmodels.formula.api as smf

pl.style.use('fivethirtyeight')
%pylab inline
```

```python
destinationZip = pd.read_csv("http://cosmo.nyu.edu/~fb55/data/destinationZIP.csv",index_col="Unnamed: 0")
```

```python
departureZip = pd.read_csv("http://cosmo.nyu.edu/~fb55/data/departureZIP.csv", 
                           index_col="Unnamed: 0")
```

```python
departureZip.head()
```

```python
df_dep = departureZip.groupby(departureZip.ZIPCODE).count()['Start_Lon']
df_dep.head()
```

```python
df_des = destinationZip.groupby(destinationZip.ZIPCODE).count()['Start_Lon']
df_des.head()
```

```python
df = pd.concat([df_dep, df_des] , axis=1)
```

```python
df = df.dropna()
df
```

```python
asthma = pd.read_csv('asthma.csv')
asthma.head()
```

```python
df.columns = ['DEP_COUNT', 'DES_COUNT']
df.head()
```

```python
df_new = df.reset_index()
```

```python
df_new.head()
```

```python
df_new.ZIPCODE.astype(int)
```

```python
asthma = asthma.rename(columns = {'Zip Code':'ZIPCODE'})
```

```python
asthma.columns
```

```python
asthma.ZIPCODE.astype(int)
```

```python
df_final = pd.merge(asthma, df_new, on='ZIPCODE')
```

```python
df_final.head()
```

```python
short = df_final[['ZIPCODE','DEP_COUNT', 'DES_COUNT','Discharge Rate']]
```

```python
short.head()
```

```python
short = short.dropna()
```

```python
feat = short['DES_COUNT']

targ = short['Discharge Rate']
```

```python
plt.scatter(feat, targ)
slope, intercept, r_value, p_value, std_err = stats.linregress(feat, targ)

plt.plot(feat, intercept + slope*feat, color='firebrick')
plt.xlabel('Number of Trips Ended in Zipcode')
plt.ylabel('Hospital discharge rate (per 10,000 population)')
plt.text(8, 45, "y = %.2f + %.2f*x" % (intercept, slope), fontsize=14)
plt.title('Asthma v. Taxi Trip Density(as destination) (corr=%.2f)' %np.corrcoef(feat, targ)[0][1])
plt.show()
```

```python
feat = short['DEP_COUNT']

targ = short['Discharge Rate']
```

```python
plt.scatter(feat, targ)
slope, intercept, r_value, p_value, std_err = stats.linregress(feat, targ)

plt.plot(feat, intercept + slope*feat, color='firebrick')
plt.xlabel('Number of Trips Started in Zipcode')
plt.ylabel('Hospital discharge rate (per 10,000 population)')
plt.text(8, 45, "y = %.2f + %.2f*x" % (intercept, slope), fontsize=14)
plt.title('Asthma v. Taxi Trip Density(as departure) (corr=%.2f)' %np.corrcoef(feat, targ)[0][1])
plt.show()
```
