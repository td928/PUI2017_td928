```python
SQL_SOURCE = 'https://td928-nyu.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

# Task 3

## Get station locations in web mercator projection

```python
q8 = '''
SELECT
    CDB_TransformToWebmercator(CDB_LatLng(
        start_station_latitude,
        start_station_longitude
        )
) as the_geom_webmercator, MIN(cartodb_id) as cartodb_id
FROM citibike
GROUP BY
      start_station_latitude,
      start_station_longitude
'''
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(q8)), sep=',')
```

## Get station locations in web mercator projection along with their ids and the number of trips started in each stations

```python
q9 = '''
SELECT
    start_station_id,
    CDB_TransformToWebmercator(CDB_LatLng(
        start_station_latitude,
        start_station_longitude
        )
) as the_geom_webmercator, MIN(cartodb_id) as cartodb_id, COUNT(tripduration) as numtrips
FROM citibike
GROUP BY
      start_station_id,
      start_station_latitude,
      start_station_longitude
'''
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(q9)), sep=',')
```

## but only limited to stations within 500m radius of Time Square

```python
q10 = '''
SELECT
    start_station_id,
    CDB_TransformToWebmercator(CDB_LatLng(
        start_station_latitude,
        start_station_longitude
        )
) as the_geom_webmercator, MIN(cartodb_id) as cartodb_id, COUNT(tripduration) as numtrips
FROM citibike
WHERE
     ST_DWithin(CDB_LatLng(
         start_station_latitude,
         start_station_longitude
)::geography, CDB_LatLng(40.7577,-73.9857)::geography, 500)
GROUP BY
      start_station_id,
      start_station_latitude,
      start_station_longitude
'''
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(q10)), sep=',')
```

# Task 4

## find the trip with longest average duration on weekends and within 500m of Times Square

```python
q11 = '''
SELECT
    start_station_id,
    CDB_TransformToWebmercator(CDB_LatLng(
      start_station_latitude,
start_station_longitude)
) as the_geom_webmercator, MIN(cartodb_id) as cartodb_id, AVG(tripduration) as avg_duration
FROM citibike
WHERE
     ST_DWithin(
         CDB_LatLng(
              start_station_latitude,
         start_station_longitude
)::geography, CDB_LatLng(40.7577,-73.9857)::geography, 500)
AND
extract(DOW FROM starttime) IN (0,6)
GROUP BY
      start_station_id,
      start_station_latitude,
      start_station_longitude
ORDER BY avg_duration DESC LIMIT 1
'''
```

```python
pd.read_csv(StringIO.StringIO(queryCartoDB(q11)), sep=',')
```
